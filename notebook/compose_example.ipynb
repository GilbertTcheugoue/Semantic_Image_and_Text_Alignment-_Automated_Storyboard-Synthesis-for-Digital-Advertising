**To create a storyboard that combines images both horizontally and vertically**

```python
from PIL import Image
import matplotlib.pyplot as plt

class StoryBoard:
    def __init__(self, images):
        self.images = images

    @classmethod
    def create_storyboard(cls, images):
        if not images or not all(isinstance(img, Image.Image) for img in images):
            raise ValueError("A valid list of PIL Image objects is required.")
        return cls(images)

    def combine_horizontal(self):
        """Combine images horizontally."""
        widths, heights = zip(*(img.size for img in self.images))
        total_width = sum(widths)
        max_height = max(heights)

        combined_image = Image.new('RGB', (total_width, max_height))
        x_offset = 0

        for img in self.images:
            combined_image.paste(img, (x_offset, 0))
            x_offset += img.width

        return combined_image

    def combine_vertical(self):
        """Combine images vertically."""
        widths, heights = zip(*(img.size for img in self.images))
        max_width = max(widths)
        total_height = sum(heights)

        combined_image = Image.new('RGB', (max_width, total_height))
        y_offset = 0

        for img in self.images:
            combined_image.paste(img, (0, y_offset))
            y_offset += img.height

        return combined_image

    def show_combined_horizontal(self):
        combined_image = self.combine_horizontal()
        plt.imshow(combined_image)
        plt.axis('off')
        plt.title('Combined Horizontally')
        plt.show()

    def show_combined_vertical(self):
        combined_image = self.combine_vertical()
        plt.imshow(combined_image)
        plt.axis('off')
        plt.title('Combined Vertically')
        plt.show()

# Example usage
if __name__ == "__main__":
    # Load your images
    images = [
        Image.open('/content/drive/MyDrive/Week12/adludio storyboard examples/Adludio-Detran_RS-[BR]-[LIVE]-Carnival_24-User_Choice_Quiz-FS-Version_2-Taxi.png'),
        Image.open('/content/drive/MyDrive/Week12/adludio storyboard examples/Adludio-Volvo-[UK]-[RFP]-Volvo_Vehicle_Electrification_XC40-Tap-FS-Version_2_AJ.png'),
        Image.open('/content/drive/MyDrive/Week12/adludio storyboard examples/Adludio-google-[UK]-[RFP]-brand_saftey-swipe_left_or_right-FS-v1.png')
    ]

    # Create the storyboard
    storyboard = StoryBoard.create_storyboard(images)

    # Show combined images
    storyboard.show_combined_horizontal()
    storyboard.show_combined_vertical()
```
